```python
import pandas as pd
import matplotlib as plt
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv("marathon_athlete_events (1).csv")
df
```

### 1.Определить общее количество спортсменов за всю историю Игр.


```python
df['Name'].nunique()
```

### 2.Какие Олимпийские игры были самыми масштабными по числу спортсменов?


```python
plt.subplots(figsize=(15, 5))
plt.xticks(fontsize=12, rotation=45)
# sns.color_palette("ch: s=-.2,r=.6", as_cmap=True)
sns.barplot(data = df.groupby('Games')['Name'].count().reset_index(),x = 'Games',y = 'Name',palette='crest')
# to_frame(name='count_sportsmens')
```

### 3.На каких Олимпийских играх (зимних и летних) было завоевано больше всего золотых медалей?

```python
gold = df[df.Medal == "Gold"]
plt.subplots(figsize=(15, 5))
plt.xticks(fontsize=12, rotation=45)
sns.barplot(data=gold.groupby('Games')['Medal'].count().reset_index(), x='Games', y='Medal',palette='crest')
```

### 4.Какие команды завоевали призовые медали в хоккее на льду во время летних Олимпийских игр, где и в каком году?

```python
pd.set_option('display.max_rows', None)
sport = df[(df.Sport == "Ice Hockey") & (df.Season=='Summer')]
sport.groupby(['Games','Team','City']).agg({'Sport':['count']})
```

### 5.Выигрывали ли спортсмены из Фиджи золотые медали?

```python
df1 = df[(df.region == 'Fiji') & (df.Medal=="Gold")]
df1
```

### 6. Кто был самым старым олимпийцем за всю историю?


```python
plt.subplots(figsize=(15, 3))
plt.xticks(fontsize=9, rotation=90)
sns.barplot(data=df.groupby('Age')['Name'].max().reset_index(), x='Name', y='Age', palette='crest')
```

### 7.На каких Олимпийских играх (зимних и летних) был самый низкий (в среднем) рост среди спортсменов?

```python
plt.subplots(figsize = (15,6))
plt.xticks(fontsize = 10,rotation = 45)
sns.barplot(data = df.groupby('Games')['Height'].mean().reset_index(),x = 'Games', y = 'Height',palette='hls')
```
