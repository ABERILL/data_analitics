### Задача1

##### Пусть дана выборка x1, . . . ,xn из равномерного распределения на отрезке[a;θ]. Найдём оценку для неизвестного параметра θ методом моментов и методом максимального правдоподобия. Исследуем полученные оценки на несмещенность и состоятельность.Реализуем эту задачу:
1. сгенерируем выборку из равномерного распределения на [a;θ] размера n= 10, 100, 1000, 10 000;
2. найдём значения полученных оценок;
3. выведим отклонения полученных оценок от параметра θ. Что происходит с ростом n?

```python
a = 28
b = a+a/2
size = [10,100,1000,10000][3]
uniform = np.random.uniform(low = a,high = b,size=size)
```

Используем метод моментов

```python
print("Истинное значение параметра: ", b)
#Так как у нас один параметр, то рассмотрим момент первого порядка,то есть математическое ожидание EX.
#EX = (28+z)/2 , откуда мы и получаем уравнение на неизвестный параметр. Вместо теоретического момента EX подставим эмпирический момент - выборочное среднее x.
a_moment = sum(uniform)#Суммируем значения выборки
x = (1/size)*a_moment #Находим выборочное среднее
z_moment = (x*2)-28
print("Значение метода моментов: ",z_moment)
```

Видим,что последнее значение максимально приближено к истинному значению. То есть ясно, что при увеличении выборки наше значение стремится к истинному, следовательно наша оценка состоятельна.

Метод максимального правдоподобия

```python
print("Истинное значение параметра: ", b)
a_pravda = a+ss.uniform.fit(uniform)[0]/2
print("Значение метода максимального правдоподобия: ",a_pravda)
```

```python
print("Смещение: ", sum(uniform)/size-b)#смещение – это разность между математическим ожиданием оценки и истинным значением оцениваемого параметра, а несмещенная оценка – это оценка, имеющая нулевое смещение
```

Видим, что оценка смещённая.

```python
print("Отклонение метода моментов: ",{(b-z_moment) * (1 if b-z_moment>0 else -1)})
print("Отклонение метода максимального правдоподобия: ",{(b-a_pravda) * (1 if b-a_pravda>0 else -1)})
```

### Задача 3

##### Пусть дана реализация выборки x1, . . . ,xn из нормального распределения N(θ1,θ2). Найдём оценки для неизвестных параметров θ1,θ2 методом моментов и методом максимального правдоподобия. Являются ли оценки θ1 несмещенными и состоятельными? Являются ли оценки θ2 несмещенными? Сделаем предположение о состоятельности оценок на θ2. Реализуем эту задачу:
1. сгенерируем выборку из нормального распределения N(θ1,θ2) размера n= 10, 100, 1000, 10 000;
2. найдём значения полученных оценок;
3. выведим отклонения полученных оценок от параметровθ1,θ2. Что происходит с ростом n?

```python
m = 28
n = 1+m/2
size = [10,100,1000,10000][3]
norm = ss.norm.rvs(loc = m,scale = n,size=size)
```

Метод максимального правдоподобия

```python
def L(args): return - \
    np.prod([ss.norm.pdf(xi, args[0], args[1]) for xi in norm])

#берём логарифм от правдоподобия


def logL(args): return - \
    np.sum([ss.norm.logpdf(xi, args[0], args[1]) for xi in norm])
```

```python
from scipy.optimize import minimize
x0 = [0, 1]
# (-inf, inf) для первого параметра (мат.ож), (0, inf)  для второго (дисперсии)
bounds = [(None, None), (0, None)]
res = minimize(L, x0, bounds=bounds)
logres = minimize(logL, x0, bounds=bounds)

print(res.x)
print(logres.x)
```

Метод моментов

```python
#Для отыскания двух неизвестных параметров необходимо иметь два уравнения:
#приравняем начальный теоретический момент первого порядка и центральный теоретический момент второго порядка соответствующим эмпирическим моментам
#v1= M(Х),М1=х_vub,µ2 = D(X),m2 =D(B), получаем систему M(Х)= x_vub
#                                                       D(X) = D(B)
#Так как математическое ожидание и дисперсия нормального распределения известны в ввиде начальных параметрых, тогда получаем систему
#M(Х)=a= x_vub
#D(X)=σ**2 = D(B)
from cmath import sqrt
a = 1/size*sum(norm)#выборочное среднее
r = stats.moment(norm, moment=2, axis=0)#находим 2 момент, который соответствует дисперсии
σ = sqrt(r)
print(a)
print(σ)
```
